# Example Visualization: Total population in 2023

```python
import matplotlib.pyplot as plt
```

# Select data for 2023

```python
year = '2023'
```

```python
data1_2023 = data1[['Country Name', year]].dropna()
```

# Plot top 20 countries by population


```python
data1_2023 = data1_2023.sort_values(by=year, ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(data1_2023['Country Name'][:20], data1_2023[year][:20], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Top 20 Countries by Population in {year}')
plt.tight_layout()
plt.show()
```
